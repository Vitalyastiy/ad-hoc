```python
import matplotlib.pyplot as plt
import numpy as np

# Определим хардскиллы
skills = ['SQL', 'Python', 'Excel', 'Data Visualization', 'Работа с закзачиками', 'Понимание процессов(БА)']

# Определим уровни навыков для каждого уровня аналитика
levels = {
    'Паша Н.': [3, 1, 10, 2, 4, 3],
    'Аня Ф.': [5, 5, 10, 4, 3, 2],
    'Таня М.': [7, 7, 10, 7, 8, 8],
    'Аня Б.': [9, 9, 10, 6, 8, 7]
}

# Создадим радарную диаграмму для каждого уровня
def create_individual_radar_charts(levels, skills):
    labels = np.array(skills)
    num_skills = len(skills)

    # Установим угол для каждой оси
    angles = np.linspace(0, 2 * np.pi, num_skills, endpoint=False).tolist()
    angles += angles[:1]

    # Создадим отдельный график для каждого уровня
    for level, values in levels.items():
        fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

        # Завершим круг, добавив первый элемент в конец списка
        values += values[:1]
        
        # Отобразим уровень
        ax.fill(angles, values, alpha=0.25, label=level)
        ax.plot(angles, values, linewidth=2, linestyle='solid')

        # Настроим шкалу от 0 до 10 с шагом 1
        ax.set_ylim(0, 10)
        ax.set_yticks(np.arange(0, 11, 1))

        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(labels)

        plt.title(f'{level} - оценка скилов')
        plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
        plt.show()

create_individual_radar_charts(levels, skills)
```
