```python
import pandas as pd
```

```python
df = pd.read_excel('deal-online-v2.deal (1) (1).xlsx')
df.head(5)
```

```python
def determine_success(row):
    if row['статус заявки'] in ['DRAFT', 'CANCELED']:
        return 'DRAFT/CANC'
    else:
        return 'SUCCESS'
df['Статус'] = df.apply(determine_success, axis=1)
df.head(5)
```

```python
df['название офиса'] = df['название офиса'].fillna('Офис не заполнен')
df['id компании офиса'] = df['id компании офиса'].astype(str)
df['месяц создания'] = pd.to_datetime(df['дата создания'])
df['месяц создания'] = df['месяц создания'].dt.to_period('M')
df['дата создания'] = pd.to_datetime(df['дата создания'])
df['дата_создания'] = df['дата создания'].dt.date
df.head(5)
```

```python
df.groupby(['id компании офиса', 'название офиса', 'Статус']).agg({'ID':'count'}).reset_index()
```

```python
#кол-во по офисам
pivot_df = df.pivot_table(index=['месяц создания', 'id компании офиса', 'название офиса'],
                          columns='Статус',
                          values='ID',
                          aggfunc='count',
                          fill_value=0).reset_index()
pivot_df.head(5)
```

```python
#кол-во по компаниям(владельцы офисов)
pivot_df2 = df.pivot_table(index=['месяц создания', 'id компании офиса'],
                          columns='Статус',
                          values='ID',
                          aggfunc='nunique',
                          fill_value=0).reset_index()
pivot_df2.head(5)
```

```python
#кол-во по компаниям, детально по заказчикам
q = ['PROFESSIONAL', 'REALTOR']
pivot_df1 = df.query("`роль заказчика` ==  @q " ).pivot_table(index=['месяц создания', 'Маркетинговое название', 'ID заказчика'],
                          columns='Статус',
                          values='ID',
                          aggfunc='count',
                          fill_value=0).reset_index()
pivot_df1
```

```python
#кол-во по компаниям, детально по заказчикам
q = ['PROFESSIONAL', 'REALTOR']
pivot_df3 = df.query("`роль заказчика` ==  @q " ).pivot_table(index=['Маркетинговое название' ],
                          columns='Статус',
                          values='ID',
                          aggfunc='nunique',
                          fill_value=0).reset_index()


pivot_df3['share'] = pivot_df3['DRAFT/CANC'] /( pivot_df3['SUCCESS'] + pivot_df3['DRAFT/CANC'])
pivot_df3
```

```python
# Находим индексы строк с максимальными значениями 'дата создания' для каждого 'Маркетинговое название'
max_dates_index = df.groupby(['Маркетинговое название']).agg({'дата создания': 'max'}).reset_index()
max_dates_index
```

```python
t = pd.merge(pivot_df3, max_dates_index[['Маркетинговое название', 'дата создания']], left_on='Маркетинговое название', right_on='Маркетинговое название', how='left')
t1 = df.groupby(['Маркетинговое название',  'статус заявки']).agg({'дата создания':'max'}).reset_index()
temp = pd.merge(t, t1[['статус заявки','Маркетинговое название', 'дата создания']], left_on=['дата создания', 'Маркетинговое название' ], right_on=['дата создания', 'Маркетинговое название' ], how='left')
temp['дата создания'] = temp['дата создания'].dt.tz_localize(None)
temp
```

```python
#кол-во по компаниям, детально по заказчикам
q = ['PROFESSIONAL', 'REALTOR']
pivot_df4 = df.query("`роль заказчика` ==  @q " ).pivot_table(index=['месяц создания', 'Маркетинговое название'],
                          columns='Статус',
                          values='ID',
                          aggfunc='count',
                          fill_value=0).reset_index()
pivot_df4
```

```python
pivot_df.to_excel('t1.xlsx')
pivot_df1.to_excel('t2.xlsx')
pivot_df2.to_excel('t3.xlsx')
temp.to_excel('t4.xlsx', index=False)
pivot_df4.to_excel('t5.xlsx')
```
