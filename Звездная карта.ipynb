```python
import matplotlib.pyplot as plt
import numpy as np

#  
skills = ['SQL', 'Python', 'Excel', 'Data Visualization', 'Statistics', 'Machine Learning']

#       
levels = {
    'Junior': [3, 4, 7, 4, 3, 0],
    'Middle': [5, 6, 9, 6, 5, 1],
    'Senior': [7, 8, 10, 8, 7, 2],
    'Lead': [9, 9, 10, 10, 8, 2]
}

#   
def create_radar_chart(levels, skills):
    labels = np.array(skills)
    num_skills = len(skills)

    #     
    angles = np.linspace(0, 2 * np.pi, num_skills, endpoint=False).tolist()
    angles += angles[:1]

    #  
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

    for level, values in levels.items():
        #  ,      
        values += values[:1]
        
        #   
        ax.fill(angles, values, alpha=0.25, label=level)
        ax.plot(angles, values, linewidth=2, linestyle='solid')

    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    plt.title('Star Chart for Analyst Hard Skills')
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    plt.show()

create_radar_chart(levels, skills)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#    
tasks = ['Task A', 'Task B', 'Task C', 'Task D', 'Task E']
skills = ['SQL', 'Python', 'Excel', 'Data Visualization', 'Statistics', 'Machine Learning']

#      
# 1 -   , 2 -  , 3 -  
data = np.array([
    [3, 2, 1, 1, 3, 2],  # Task A
    [2, 3, 1, 3, 2, 1],  # Task B
    [1, 2, 3, 2, 3, 2],  # Task C
    [2, 1, 2, 3, 1, 3],  # Task D
    [3, 3, 2, 2, 2, 1],  # Task E
])

#  HeatMap
plt.figure(figsize=(10, 8))
ax = sns.heatmap(data, annot=True, fmt="d", cmap="YlGnBu", xticklabels=skills, yticklabels=tasks, cbar_kws={'label': 'Usage Frequency'})

#   
plt.xlabel('Skills')
plt.ylabel('Tasks')
plt.title('HeatMap of Tasks vs Skills Usage')

plt.show()
```
