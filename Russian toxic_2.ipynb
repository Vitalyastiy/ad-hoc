```python
import pandas as pd
from transformers import pipeline
```

```python
reviews = pd.read_excel('test.xlsx')
df = reviews.dropna()
```

Подготовим текст:

1.Удалим переносы строк.


2.Оставим только буквенный текст (удалим знаки препинания, цифры и иные символы)


3.Приведём текст к нижнему регистру.

```python
df.mark_6 = df.mark_6.str.replace('\n', ' ', regex=False)
df.mark_6 = df.mark_6.str.replace('ё', 'е', regex=False)
df.mark_6 = df.mark_6.str.replace('[^a-zA-Zа-яА-Я]+', ' ')
df.mark_6 = df.mark_6.str.lower()
```

```python
from transformers import pipeline

clf = pipeline(
    task = 'sentiment-analysis', 
    model = 'SkolkovoInstitute/russian_toxicity_classifier')

#text = ['пятигорск']
#clf(text)
classifier = pipeline("sentiment-analysis", model="tinkoff-ai/response-toxicity-classifier-base")
```

```python
def fix_addr(text):
    matches = clf(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['city']=df['mark_6'].map(fix_addr)
df
```

```python
classifier_for_sms = pipeline("sentiment-analysis", model="cointegrated/rubert-tiny-sentiment-balanced")
classifier_t = pipeline("sentiment-analysis", model="Tatyana/rubert-base-cased-sentiment-new")
classifier_vk = pipeline("sentiment-analysis", model="sismetanin/rubert-ru-sentiment-rusentiment")
classifier_emot = pipeline("sentiment-analysis", model="cointegrated/rubert-tiny2-cedr-emotion-detection")

classifier_feed = pipeline("sentiment-analysis", model="seninoseno/rubert-base-cased-sentiment-study-feedbacks-solyanka") 
#classifier_dp = pipeline("sentiment-analysis", model="blanchefort/rubert-base-cased-sentiment")
```

```python
def fix_addr(text):
    matches = classifier_t(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['Tatyana']=df['mark_6'].map(fix_addr)
```

```python
def fix_addr(text):
    matches = classifier_for_sms(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['for_sms']=df['mark_6'].map(fix_addr)
```

```python
def fix_addr(text):
    matches = classifier_vk(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['vk']=df['mark_6'].map(fix_addr)
def fix_addr(text):
    matches = classifier_emot(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['emot']=df['mark_6'].map(fix_addr)
df
```

```python
def fix_addr(text):
    matches = classifier_feed(text)
    return matches #(', '.join(f'{match or ""} {match}' for match in matches)) or None
df['feed']=df['mark_6'].map(fix_addr)
df
```

```python
df.to_excel("output.xlsx") 
```
