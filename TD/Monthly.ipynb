```python
df = pd.read_excel('Трекинг_Годовые данные NPS_Регионы МР_2021_2022.xlsx')
```

```python
df= df[['Показатель', 'Оператор', 'Категория', 'Категория_Выбор', 'Разбивка2', '2021-01-01', '2022-01-01', 'Base_2021', 'Base_2022']]
df.rename(columns={2021:"2021-01-01", 2022:"2022-01-01"}, inplace=True)
df.columns
```

```python
#переворчиваем нпс
df_unpivot = pd.melt(df, id_vars=['Показатель', 'Оператор', 'Категория', 'Категория_Выбор', 'Разбивка2'], \
    value_vars=['2021-01-01', '2022-01-01'], var_name='year', value_name='NPS')
df_unpivot
```

```python
#переворчиваем базу
df_unpivot_b = pd.melt(df, id_vars=['Показатель', 'Оператор', 'Категория', 'Категория_Выбор', 'Разбивка2'], \
    value_vars=['Base_2021', 'Base_2022'], var_name='year_cnt', value_name='subs')
df_unpivot_b
```

```python
#merge
df_full=pd.merge(df_unpivot , df_unpivot_b[['year_cnt', 'subs']], left_index= True, right_index=True, how='left') # 'on =' - для джойна по столбцу
df_full
```

```python
df_full['year']= pd.to_datetime( df_full['year'])
```

```python
df_full.info()
```

```python
df_full.to_excel('итог-2021-2022.xlsx') 
```
