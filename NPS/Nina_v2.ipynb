```python
df_ntd.columns = ['create_date', 'subs_id', 'Region', 'branch_id', 'Lifetime',
       'point_name', 'mark_6', 'nps', 'ltr', 'subs_flg_nina', 'up_price',
       'sms_ind', 'year', 'month']
df_rtd.columns = ['create_date', 'subs_id', 'Region', 'branch_id', 'Lifetime',
       'point_name', 'mark_6', 'nps', 'ltr', 'subs_flg_nina', 'up_price',
       'sms_ind', 'year', 'month']      
```

```python
total_df = pd.concat([df, df_ntd, df_rtd], axis=0, join='inner', ignore_index=True)
total_df
```

```python
total_df.point_name.value_counts()
```

```python
cut=total_df[total_df['up_price'].notnull()].copy()
cut['up_price'] = cut['up_price'].str.replace(',', '.').str.replace(' ', '')
cut['up_price'] = cut['up_price'].astype (float)
cut['quantile_ex_1'] = pd.qcut(cut['up_price'], q=3)
cut['quantile_ex_1'].value_counts()
```

```python
cut
```

```python
sum_df = pd.merge(total_df, cut[['subs_id', 'create_date','quantile_ex_1']], how='left', left_on=['subs_id', 'create_date'], right_on = ['subs_id', 'create_date']) # merge
sum_df
```
